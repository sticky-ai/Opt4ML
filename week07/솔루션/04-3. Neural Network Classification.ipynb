```python
import numpy as np
import tensorflow as tf
from sklearn.datasets import make_blobs
from matplotlib.pylab import plt
from pandas import DataFrame
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
# generate 2d classification dataset
xy, labels = make_blobs(n_samples=400, center_box=(-1,1), centers=6,cluster_std=0.1, random_state=20)
labels = labels % 2
# scatter plot, dots colored by class value
df = DataFrame(dict(x=xy[:,0], y=xy[:,1], label=labels))
colors = {1:'red', 0:'blue'}
fig, ax = plt.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key], edgecolor='k')
plt.axis('equal')
plt.grid()
plt.show()
```

```python
features = np.array([[xval, yval] for xval, yval in xy])
labels = labels.reshape(-1,1)
```

```python
def plot_model(sess, model):
    xx, yy = np.meshgrid(np.linspace(-1.5,1.5), np.linspace(-1.5,1.5))
    prediction = sess.run(model, feed_dict={x: np.array([[xxval, yyval] for xxval, yyval in zip(xx.flatten(), yy.flatten())])})
    Z = prediction.reshape(xx.shape)
    df = DataFrame(dict(x=xy[:,0], y=xy[:,1], label=labels.flatten()))
    colors = {1:'red', 0:'blue'}
    _, ax = plt.subplots(figsize=(5, 5))
    ax.contourf(xx, yy, Z, cmap='coolwarm', alpha=.9,)
    grouped = df.groupby('label')
    for key, group in grouped:
        group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key], edgecolor='k')
    ax.set_xlim([-1.3, 1.3])
    ax.set_ylim([-1.3, 1.3])
    plt.grid()
    plt.show()
```

## Neural Network 1

    .

1. Hidden Layer 1 : $h_1 = \sigma(xW_1 + b_1) $
1. Hidden Layer 2 : $h_2 = \sigma(h_1W_2 + b_2) $
1. Output Layer : $\hat{y} = \sigma(h_2W_3 + b_3)  $

 Hyper-Parameter     .

* `num_hidden1` : 10
* `num_hidden2` : 4
* `activation` : `tf.nn.sigmoid`

Hidden Layer 1, 2 Output Layer    .

1. `tf.layers.dense(input, units = , use_bias = , activation = )`  Hidden Layer 1 
1. `tf.layers.dense(input, units = , use_bias = , activation = )`  Hidden Layer 2 
1. `tf.layers.dense(input, units = , use_bias = , activation = )`  Output Layer 

```python
num_hidden1 = 10
num_hidden2 = 4
tf.set_random_seed(180410)
x = tf.placeholder(tf.float32, [None, 2])
# TODO 1
# TODO 2
# TODO 3
hidden1 = tf.layers.dense(x, units=num_hidden1, use_bias=True, activation=tf.nn.sigmoid)
hidden2 = tf.layers.dense(hidden1, units=num_hidden2, use_bias=True, activation=tf.nn.sigmoid)
model = tf.layers.dense(hidden2, units=1, use_bias=True, activation=tf.nn.sigmoid)
```

```python
y = tf.placeholder(tf.float32, [None, 1])

loss = tf.reduce_mean(-y * tf.log(model) - (1 - y) * tf.log(1 - model))

train = tf.train.AdamOptimizer().minimize(loss)
```

```python
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)

for epoch in range(10001):
    sess.run(train, feed_dict={x : features, y:labels})
    if epoch % 2000 == 0:
        curr_loss = sess.run(loss, feed_dict={x : features, y:labels})
        print(epoch, curr_loss)
        plot_model(sess, model)
```

```python
plot_model(sess, model)
```

## Neural Network 2

    .

1. Hidden Layer 1 : $h_1 = relu(xW_1 + b_1) $
1. Hidden Layer 2 : $h_2 = relu(h_1W_2 + b_2) $
1. Output Layer : $\hat{y} = \sigma(h_2W_3 + b_3)  $

 Hyper-Parameter     .

* `num_hidden1` : 10
* `num_hidden2` : 4
* `activation` : `tf.nn.relu` and `tf.nn.sigmoid`

Hidden Layer 1, 2 Output Layer    .

1. `tf.layers.dense(input, units = , use_bias = , activation = )`  Hidden Layer 1 
1. `tf.layers.dense(input, units = , use_bias = , activation = )`  Hidden Layer 2 
1. `tf.layers.dense(input, units = , use_bias = , activation = )`  Output Layer 

```python
num_hidden1 = 10
num_hidden2 = 4
tf.set_random_seed(180410)
x = tf.placeholder(tf.float32, [None, 2])
# TODO 4
# TODO 5
# TODO 6
hidden1 = tf.layers.dense(x, units=num_hidden1, use_bias=True, activation=tf.nn.relu)
hidden2 = tf.layers.dense(hidden1, units=num_hidden2, use_bias=True, activation=tf.nn.relu)
model = tf.layers.dense(hidden2, units=1, use_bias=True, activation=tf.nn.sigmoid)

y = tf.placeholder(tf.float32, [None, 1])

loss = tf.reduce_mean(-y * tf.log(model) - (1 - y) * tf.log(1 - model))

train = tf.train.AdamOptimizer().minimize(loss)
```

```python
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)

for epoch in range(4001):
    sess.run(train, feed_dict={x : features, y:labels})
    if epoch % 500 == 0:
        curr_loss = sess.run(loss, feed_dict={x : features, y:labels})
        print(epoch, curr_loss)
        plot_model(sess, model)
```
